```python
import pandas as pd
from datetime import datetime, timedelta

# Function to parse date and time in the specified format
def parse_datetime(dt_str):
    return datetime.strptime(dt_str, "%m/%d/%Y %I:%M %p")
file_path = 'Assignment_Timecard.xlsx'
df = pd.read_excel(file_path)
df
```

```python

def calculate_time_difference(start_time, end_time):
    return (end_time - start_time).total_seconds() / 3600
```

```python
def helper(file_path):
    df = pd.read_excel(file_path)
    consecutive_days = 7
    consecutive_days_mask = df['Time'].diff().dt.days.eq(1).rolling(consecutive_days).sum() == consecutive_days - 1
    employees_7_consecutive_days = df[consecutive_days_mask]
    print("Employees who have worked for 7 consecutive days:")
    print(employees_7_consecutive_days[['Employee Name', 'Position ID']])
    print("\n")

    df['time_difference'] = df['Time'].diff().fillna(timedelta(seconds=0))
    time_between_shifts_mask = (df['time_difference'] > timedelta(hours=1)) & (df['time_difference'] < timedelta(hours=10))
    employees_between_shifts = df[time_between_shifts_mask]
    print("Employees who have less than 10 hours of time between shifts but greater than 1 hour:")
    print(employees_between_shifts[['Employee Name', 'Position ID']])
    print("\n")

    max_hours_in_single_shift = 14
    long_shift_mask = df.apply(lambda row: calculate_time_difference(row['Time'], row['Time Out']) > max_hours_in_single_shift, axis=1)
    employees_long_shift = df[long_shift_mask]
    print("Employees who have worked for more than 14 hours in a single shift:")
    print(employees_long_shift[['Employee Name', 'Position ID']])
    print("\n")

```

```python
file_path = 'Assignment_Timecard.xlsx'
helper(file_path)
```
